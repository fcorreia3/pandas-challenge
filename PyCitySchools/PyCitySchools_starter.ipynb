```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data.head()
```

```python
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_data_complete = school_data_complete.set_index("Student ID")
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
### Calculate the total number of schools ###

#school_data_complete_grouped1 = school_data_complete.groupby("School ID")
school_data_complete_grouped2 = school_data_complete.groupby("school_name")

#schools_total1 = school_data_complete_grouped["School ID"].count()
schools_total2 = school_data_complete_grouped2["school_name"].count()

#print(schools_total1, schools_total2)

total_schools = schools_total2.count()
print(f"The total number of schools { total_schools }")

```

```python
### Calculate the total number of students ###

school_data_complete["student_name"].describe()
total_students = school_data_complete["student_name"].count()
print(f"The total number of students is { total_students }")
```

```python
### Calculate the total budget ###

school_data_complete_grouped2 = school_data_complete.groupby("school_name")

#schools_total2 = school_data_complete_grouped2["school_name"].count()
school_data_complete_grouped2["budget"].describe()
school_data_complete_grouped2["budget"].max()

total_budget = sum(school_data_complete_grouped2["budget"].max())

print(f"The total budget is ${ total_budget }")
```

```python
### Calculate the average math score ###

school_data_complete["math_score"].describe()
avg_math = school_data_complete["math_score"].mean()

print(f"The average math score is { avg_math }%")
```

```python
### Calculate the average reading score ###

school_data_complete["reading_score"].describe()
avg_reading = school_data_complete["reading_score"].mean()

print(f"The average reading score is { avg_reading }%")
```

```python
### Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2 ###

avg_score = (avg_math + avg_reading)/2
avg_score

print(f"The overall average score is { avg_score }%")
```

```python
### Calculate the percentage of students with a passing math score (70 or greater) ###

pass_math_stu = school_data_complete.loc[school_data_complete["math_score"] >= 70, "student_name"]
#print(pass_math_stu)

pass_math_stu_count = pass_math_stu.count().astype(float)
#print(pass_math_stu_count)

pass_math = (pass_math_stu_count/total_students) * 100
print(f"The percentage of students with a passing math score is { pass_math }%")

```

```python
### Calculate the percentage of students with a passing reading score (70 or greater) ###

pass_reading_stu = school_data_complete.loc[school_data_complete["reading_score"] >= 70, "student_name"]

pass_reading_stu_count = pass_reading_stu.count().astype(float)

pass_reading = (pass_reading_stu_count/total_students) * 100
print(f"The percentage of students with a passing reading score is { pass_reading }%")
```

```python
### Create a dataframe to hold the above results ###

#results_list = [total_schools, total_students, total_budget, avg_math, avg_reading, avg_score, pass_math, 
#               pass_reading]
#results_labels = ["Total number schools", "Total number students", "Total budget", "Average math score", 
#                  "Average reading score", "Overall average score", "% passing math score", 
#                  "% passing reading score"]

#results_data = pd.DataFrame(results_list,results_labels)

results_data = pd.DataFrame({"Total number schools": [total_schools],
                              "Total number students": [total_students],
                              "Total budget": [total_budget],
                              "Average math score": [avg_math],
                              "Average reading score": [avg_reading],
                              "Overall average score": [avg_score],
                              "% passing math score": [pass_math],
                              "% passing reading score": [pass_reading]})

results_data
```

```python
### Optional: give the displayed data cleaner formatting ###

results_data = results_data.round(2)
results_data["Total number students"] = results_data["Total number students"].map("{0:,.0f}".format)
results_data["Total budget"] = results_data["Total budget"].map("${0:,.0f}".format)
results_data["Average math score"] = results_data["Average math score"].map("{0:,.2f}%".format)
results_data["Average reading score"] = results_data["Average reading score"].map("{0:,.2f}%".format)
results_data["Overall average score"] = results_data["Overall average score"].map("{0:,.2f}%".format)
results_data["% passing math score"] = results_data["% passing math score"].map("{0:,.2f}%".format)
results_data["% passing reading score"] = results_data["% passing reading score"].map("{0:,.2f}%".format)
results_data
```

```python
### Create an overview table that summarizes key metrics about each school
#    and create a dataframe to hold the results ###
```

```python
# School Name

schools = pd.DataFrame(school_data_complete["school_name"].value_counts())
schools
                            
schools.reset_index(inplace=True)
schools.columns = ["Schools names", "Count"]
schools
```

```python
# School Type

school_type = pd.DataFrame(school_data_complete.groupby("school_name")["type"].max())
school_type.reset_index(inplace=True)
school_type.columns = ["Schools names", "School type"]
school_type
```

```python
merged_df = pd.merge(schools, school_type, on="Schools names")
merged_df.head()
```

```python
# Total Students

students = pd.DataFrame(school_data_complete.groupby("school_name")["student_name"].count())
students.reset_index(inplace=True)
students.columns = ["Schools names", "student_name"]
students
```

```python
merged_df = pd.merge(merged_df, students, on="Schools names")
merged_df.head()
```

```python
merged_df = merged_df.rename(columns={"student_name":"Number of students"})
merged_df.head()
```

```python
merged_df = merged_df[['Schools names', 'School type', "Number of students"]]
merged_df.head()
```

```python
# Total School Budget

budget = pd.DataFrame(school_data_complete.groupby("school_name")["budget"].max())
budget.reset_index(inplace=True)
budget.columns = ["Schools names", "Budget"]
budget
```

```python
merged_df = pd.merge(merged_df, budget, on="Schools names")
merged_df.head()
```

```python
# Per Student Budget

budget_per_student = merged_df["Budget"]/merged_df["Number of students"]
merged_df["Budget per student"] = budget_per_student
merged_df.head()
```

```python
# Average Math Score

avg_math_score = pd.DataFrame(school_data_complete.groupby("school_name")["math_score"].mean())
avg_math_score.reset_index(inplace=True)
avg_math_score.columns = ["Schools names", "Average math score"]
avg_math_score
```

```python
merged_df = pd.merge(merged_df, avg_math_score, on="Schools names")
merged_df.head()
```

```python
# Average Reading Score

avg_reading_score = pd.DataFrame(school_data_complete.groupby("school_name")["reading_score"].mean())
avg_reading_score.reset_index(inplace=True)
avg_reading_score.columns = ["Schools names", "Average reading score"]
avg_reading_score
```

```python
merged_df = pd.merge(merged_df, avg_reading_score, on="Schools names")
merged_df.head()
```

```python
# % Passing Math

passing_math_count = school_data_complete.loc[(school_data_complete["math_score"] >= 70)].groupby("school_name").count()["math_score"]
students_per_school = school_data_complete.groupby("school_name").count()["math_score"]
school_math_pass = []
school_math_pass = (passing_math_count/students_per_school)*100
school_math_pass
pd_math_pass = pd.DataFrame(school_math_pass)
```

```python
pd_math_pass.reset_index(inplace=True)
pd_math_pass.columns = ["Schools names", "% Passing math"]
pd_math_pass.head()
```

```python
merged_df = pd.merge(merged_df, pd_math_pass, on="Schools names")
merged_df.head()
```

```python
# % Passing Reading

passing_read_count = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)].groupby("school_name").count()["reading_score"]
school_reading_pass = (passing_read_count/students_per_school)*100
school_reading_pass
pd_read_pass = pd.DataFrame(school_reading_pass)
```

```python
pd_read_pass.reset_index(inplace=True)
pd_read_pass.columns = ["Schools names", "% Passing reading"]
pd_read_pass.head()
```

```python
merged_df = pd.merge(merged_df, pd_read_pass, on="Schools names")
merged_df.head()
```

```python
# Overall Passing Rate (Average of the above two)

overall_pass_rate = (school_math_pass+school_reading_pass)/2
overall_pass_rate
pd_overall_pass = pd.DataFrame(overall_pass_rate)
```

```python
pd_overall_pass.reset_index(inplace=True)
pd_overall_pass.columns = ["Schools names", "Overall passing rate"]
pd_overall_pass.head()
```

```python
## FINAL SUMMARY TABLE ##

merged_df = pd.merge(merged_df, pd_overall_pass, on="Schools names")
merged_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
merged_df['Budget per student'].min()
```

```python
# Creating the bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
merged_df["Spending buckets"] = pd.cut(merged_df['Budget per student'], spending_bins, group_names)
merged_df.head()
```

```python
merged_df_grouped  = merged_df.groupby('Spending buckets')
```

```python
# Average Math Score
avg_math_per_bucket = merged_df_grouped['Average math score'].mean()

# Average Reading Score
avg_read_per_bucket = merged_df_grouped['Average reading score'].mean()

# % Passing Math
pass_math_per_bucket = merged_df_grouped['% Passing math'].mean()

# % Passing Reading
pass_read_per_bucket = merged_df_grouped['% Passing reading'].mean()

# Overall Passing Rate (Average of the above two)
pass_rate_per_bucket = merged_df_grouped['Overall passing rate'].mean()
```

```python
# Put the results in DataFrame           
scores_by_spending = pd.DataFrame({"Average math score": avg_math_per_bucket, 
                                   "Average reading score": avg_read_per_bucket,
                                   "% Passing math": pass_math_per_bucket,
                                   "% Passing reading": pass_read_per_bucket,
                                   "Overall passing rate": pass_rate_per_bucket})

scores_by_spending
```

```python
# It is evident again here that big budgets do not drive results translated into passing scores in math and 
#reading. Indeed, from this analysis, the relationship is negative.
```
